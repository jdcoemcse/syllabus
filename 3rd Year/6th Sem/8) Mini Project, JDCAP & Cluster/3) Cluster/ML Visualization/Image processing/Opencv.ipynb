```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
img = cv.imread('cat_img.jpg')
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

tx, ty = 50, 30  # shift 50px right, 30px down
matrix = np.float32([[1, 0, tx], [0, 1, ty]])
translated = cv.warpAffine(img_rgb, matrix, (img.shape[1], 
                                             img.shape[0]))
plt.imshow(translated)
plt.title("Translated Image")
plt.axis('off')
plt.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

# Load image and convert to RGB
img = cv.imread('cat_img.jpg')
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Split channels
R, G, B = cv.split(img_rgb)

# Style
plt.style.use('seaborn-v0_8-deep')

# Create 3x2 subplot: 3 channels × (Image + Histogram)
fig, axs = plt.subplots(3, 2, figsize=(14, 10))
fig.suptitle('RGB Channels and Their Histograms', fontsize=20, fontweight='bold')

# Red Channel
axs[0, 0].imshow(R, cmap='Reds')
axs[0, 0].set_title('Red Channel Image', fontsize=14)
axs[0, 0].axis('off')
axs[0, 1].hist(R.ravel(), bins=256, color='red', alpha=0.7)
axs[0, 1].set_title('Red Channel Histogram', fontsize=14)
axs[0, 1].set_xlim([0, 256])
axs[0, 1].grid(True)

# Green Channel
axs[1, 0].imshow(G, cmap='Greens')
axs[1, 0].set_title('Green Channel Image', fontsize=14)
axs[1, 0].axis('off')
axs[1, 1].hist(G.ravel(), bins=256, color='green', alpha=0.7)
axs[1, 1].set_title('Green Channel Histogram', fontsize=14)
axs[1, 1].set_xlim([0, 256])
axs[1, 1].grid(True)

# Blue Channel
axs[2, 0].imshow(B, cmap='Blues')
axs[2, 0].set_title('Blue Channel Image', fontsize=14)
axs[2, 0].axis('off')
axs[2, 1].hist(B.ravel(), bins=256, color='blue', alpha=0.7)
axs[2, 1].set_title('Blue Channel Histogram', fontsize=14)
axs[2, 1].set_xlim([0, 256])
axs[2, 1].grid(True)

# Layout adjustment
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

# Set a beautiful plotting style
plt.style.use('seaborn-v0_8-deep')

# Load and convert the image
img = cv.imread('cat_img.jpg')
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Create the figure
fig, axs = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle('Original Image and RGB Histogram', fontsize=18, fontweight='bold')

# Left: Image display
axs[0].imshow(img_rgb)
axs[0].set_title('Original Image', fontsize=14)
axs[0].axis('off')

# Right: RGB Histogram
colors = ('r', 'g', 'b')
for i, color in enumerate(colors):
    hist = cv.calcHist([img], [i], None, [256], [0, 256])
    axs[1].plot(hist, color=color, linewidth=2, alpha=0.7, label=f'{color.upper()} Channel')

axs[1].set_title('RGB Histogram', fontsize=14)
axs[1].set_xlabel('Pixel Intensity', fontsize=12)
axs[1].set_ylabel('Frequency', fontsize=12)
axs[1].legend(loc='upper right')
axs[1].grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load classifiers
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 
                    'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +
                                    'haarcascade_eye.xml')

# Load image (replace with a frame from webcam if needed)
img = cv2.imread('Sharukh.jpeg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect faces
faces = face_cascade.detectMultiScale(gray, 1.3, 5)

for (x, y, w, h) in faces:
    # Draw rectangle around face
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 255, 0), 2)

    # ROI for eyes
    roi_gray = gray[y:y + h, x:x + w]
    roi_color = img[y:y + h, x:x + w]

    # Detect eyes
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 127, 255), 2)

# Convert BGR to RGB for matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display using matplotlib
plt.figure(figsize=(8, 6))
plt.imshow(img_rgb)
plt.title('Face & Eye Detection')
plt.axis('off')
plt.show()
```
