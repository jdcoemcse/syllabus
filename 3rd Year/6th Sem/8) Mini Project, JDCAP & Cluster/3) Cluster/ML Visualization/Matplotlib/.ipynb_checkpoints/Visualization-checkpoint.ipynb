```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['5', '15', '30', '45', '60']
values1 = [19, 22, 35, 38, 40]
values2 = [12, 28, 31, 33, 35]

# Setting up positions for bars
x = np.arange(len(categories))  # the label locations

# Width of the bars
width = 0.25  

# Creating the bar plots
fig, ax = plt.subplots(figsize=(15, 6))  # Enlarged figure size
bars1 = ax.bar(x - width/2, values1, width, label='EEPPM')
bars2 = ax.bar(x + width/2, values2, width, label=' Proposed')

# Adding labels and title
ax.set_xlabel('MOBILE NODES')
ax.set_ylabel('END TO END DELAY MICROSECONDS')
#ax.set_title('Bar Plot with Two Categories')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend(ncol=2, loc='upper center')
plt.ylim(0,50)
# Display the plot
plt.show()
```
